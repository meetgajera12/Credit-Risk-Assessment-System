### `Organization-wise Analysis`

The original `ORGANIZATION_TYPE` feature contained a large number of unique categories, making visualizations difficult to interpret. Therefore, similar organization types were grouped into **12 broader organization groups (`ORG_GROUP`)** to improve readability and simplify analysis while preserving meaningful business information.

### Key Observations

- **Default Rate**
  - Employees in **Banking & Finance** have the **lowest default rate (~5.3%)**.
  - **Construction** has the **highest default rate (~10.5%)**, followed by **Trade & Self-employed**, **Hospitality**, and **Transport**.
  - This suggests that repayment behavior varies across employment sectors.

- **Income Distribution**
  - Applicants working in **Banking & Finance** report the highest average annual income.
  - **Education**, **Hospitality**, and **Other** sectors have comparatively lower average incomes.
  - Overall, income differences between organization groups are moderate.

- **Credit Amount**
  - The average credit amount is relatively consistent across organization groups.
  - **Construction** and **Government** employees receive slightly higher average credit amounts, while **Hospitality** receives comparatively lower credit amounts.

- **Annuity Amount**
  - Average annuity values closely follow the trend observed in credit amounts.
  - Applicants from **Construction** have the highest average annuity, whereas the **Other** category has the lowest.

### Conclusion

Grouping the original organization categories into broader groups improves interpretability without losing important business insights. The analysis indicates that employment sector is associated with differences in **default rate**, **income**, and, to a lesser extent, **credit and annuity amounts**, making `ORG_GROUP` a potentially useful feature for credit risk prediction.
