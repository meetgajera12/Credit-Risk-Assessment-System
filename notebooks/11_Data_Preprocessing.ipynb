## Encoding

### Categorical Feature Encoding

The dataset contains a mixture of **binary, ordinal, and nominal categorical features**. Different encoding strategies were applied based on the nature of each feature.

- **Binary categorical features** (`CODE_GENDER`, `FLAG_OWN_CAR`, `FLAG_OWN_REALTY`, `NAME_CONTRACT_TYPE`) were encoded as **0/1** since they contain only two categories.
- **`NAME_EDUCATION_TYPE`** was encoded using a **manual ordinal mapping** because education levels have a natural progression from lower to higher education.
- The remaining categorical features (`NAME_INCOME_TYPE`, `OCCUPATION_TYPE`, `NAME_FAMILY_STATUS`, `NAME_HOUSING_TYPE`, and `ORG_GROUP`) were encoded using **Ordinal Encoding**. Although these features are nominal, ordinal encoding was chosen because the final models (Random Forest, XGBoost, LightGBM, and CatBoost) are tree-based and can effectively utilize integer-encoded categorical features without requiring one-hot encoding.
- `OrdinalEncoder` was configured to assign **-1** to unseen categories, ensuring robust handling of new values during inference.

```python
binary_map = {
    'CODE_GENDER':{'F':0,'M':1},
    'FLAG_OWN_CAR':{'N':0,'Y':1},
    'FLAG_OWN_REALTY':{'N':0,'Y':1},
    'NAME_CONTRACT_TYPE':{'Revolving loans':0,'Cash loans':1}
}

for col, mapping in binary_map.items():
    X_train[col] = X_train[col].map(mapping)
    X_test[col] = X_test[col].map(mapping)
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer


ordinal_col = ['NAME_EDUCATION_TYPE']
education_order = [[
    'Lower secondary',
    'Secondary / secondary special',
    'Incomplete higher',
    'Higher education',
    'Academic degree'
]]

nominal_col = [
    'NAME_INCOME_TYPE', 'OCCUPATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE', 'ORG_GROUP'
]


preprocessor = ColumnTransformer(
        transformers=[
            (
                'ordinal',
                OrdinalEncoder(
                    categories=education_order,
                    handle_unknown='use_encoded_value',
                    unknown_value=-1
                ),
                ordinal_col
            ),
            (
                'nominal',
                OrdinalEncoder(
                    handle_unknown='use_encoded_value',
                    unknown_value=-1
                ),
                nominal_col
            )
        ],
        remainder='passthrough',
        verbose_feature_names_out=False
).set_output(transform='pandas')


X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)
```

## Base Model

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

model1 = RandomForestClassifier(class_weight='balanced',random_state=42)
model1.fit(X_train,y_train)

model2 = XGBClassifier(scale_pos_weight=(y_train.value_counts()[0]/y_train.value_counts()[1]), random_state=42)
model2.fit(X_train,y_train)

model3 = LGBMClassifier(class_weight='balanced',random_state=42)
model3.fit(X_train,y_train)

model4 = CatBoostClassifier(auto_class_weights='Balanced',verbose=0,random_state=42)
model4.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

# prediction
rf_pred = model1.predict(X_test)
xgb_pred = model2.predict(X_test)
lgbm_pred = model3.predict(X_test)
cat_pred = model4.predict(X_test)

# probabilities
rf_prob = model1.predict_proba(X_test)[:,1]
xgb_prob = model2.predict_proba(X_test)[:,1]
lgbm_prob = model3.predict_proba(X_test)[:,1]
cat_prob = model4.predict_proba(X_test)[:,1]


models = {
    "RandomForest":(rf_pred,rf_prob),
    "XGBoost":(xgb_pred,xgb_prob),
    "LightGBM":(lgbm_pred,lgbm_prob),
    "CatBoost":(cat_pred,cat_prob)
}

result = []
for name, (pred,prob) in models.items():
        result.append({
                "Model":name,
                "Accuracy" :  accuracy_score(y_test,pred),
                "Precision":  precision_score(y_test,pred),
                "Recall"   :  recall_score(y_test,pred),
                "F1 Score" :  f1_score(y_test,pred),
                "ROC-AUC"  :  roc_auc_score(y_test,prob)
        })

result = pd.DataFrame(result)
result.sort_values("ROC-AUC", ascending=False)        
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

for name, (pred,prob) in models.items():
    fpr, tpr, _ = roc_curve(y_test,prob)
    roc_auc = auc(fpr,tpr)
    
    plt.plot(fpr, tpr, label=f'{name} (AUC={roc_auc:.3f})')

plt.plot([0,1],[0,1],"--", color='gray')
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid(True)

plt.show()
```

## ROC Curve Comparison

The Receiver Operating Characteristic (ROC) curve was used to evaluate the classification performance of all baseline models. The ROC-AUC score measures the model's ability to distinguish between default and non-default applicants, where a value closer to 1 indicates better discrimination.

### Observations

- **LightGBM** achieved the highest ROC-AUC score (**0.738**), indicating the best overall classification performance among the evaluated models.
- **CatBoost** closely followed with an ROC-AUC of **0.733**.
- **XGBoost** achieved an ROC-AUC of **0.721**, performing competitively but slightly below LightGBM and CatBoost.
- **Random Forest** produced the lowest ROC-AUC (**0.693**), suggesting comparatively weaker discriminative ability on this dataset.

### Conclusion

Among the baseline models, **LightGBM demonstrated the strongest predictive performance**. Therefore, it was selected for further optimization using hyperparameter tuning and model explainability techniques.

# Model Tunning

```python
import optuna
from sklearn.model_selection import cross_val_score

def objective(trial):
    n_estimator = trial.suggest_int('n_estimator',100,500)
    learning_rate = trial.suggest_float('learning_rate',0.005,0.2,log=True)
    max_depth = trial.suggest_int('max_depth',3,15)
    num_leaves = trial.suggest_int('num_leaves',20,100)
    min_child_samples = trial.suggest_int('min_child_samples',10,100)
    subsample = trial.suggest_float("subsample",0.6,1.0)
    colsample_bytree = trial.suggest_float('colsample_bytree',0.6,1.0)

    model = LGBMClassifier(
        n_estimators=n_estimator,
        learning_rate=learning_rate,
        max_depth=max_depth,
        num_leaves=num_leaves,
        min_child_samples=min_child_samples,
        subsample=subsample,
        colsample_bytree=colsample_bytree,
        class_weight='balanced',
        random_state=42,
        n_jobs=-1
    )

    score = cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc', n_jobs=-1).mean()
    return score


study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)
```

```python
print(study.best_params)
print(study.best_value)
```

```python
best_model = LGBMClassifier(
    n_estimators=461,
    learning_rate=0.026155199976908773,
    max_depth=14,
    num_leaves=36,
    min_child_samples=98,
    subsample=0.7657507962678198,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

best_model.fit(X_train,y_train)
```

## SMOTE Evaluation

SMOTE was applied to the training set to address class imbalance, and the resulting model was compared with the LightGBM model trained using built-in class weighting.

### Results

| Model | ROC-AUC | Precision (Class 1) | Recall (Class 1) | F1-score (Class 1) |
|--------|--------:|--------------------:|-----------------:|-------------------:|
| LightGBM + Class Weight | **0.7405** | 0.16 | **0.66** | **0.26** |
| LightGBM + SMOTE | 0.7136 | **0.30** | 0.01 | 0.02 |

### Conclusion

Applying SMOTE did not improve model performance. Although precision for the default class increased, recall dropped dramatically, meaning the model failed to identify the vast majority of actual defaulters. The ROC-AUC score also decreased compared with the class-weighted model.

Therefore, the final model uses **LightGBM with built-in class weighting**, as it provides substantially better discrimination and default detection without introducing synthetic samples.
