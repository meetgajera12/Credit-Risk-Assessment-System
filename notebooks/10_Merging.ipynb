```python
merge_df2 = merg1.merge(
    prev,
    on='SK_ID_CURR',
    how='left'
)
```

## Post-Merge Missing Value Handling

```python
merge_df2.isnull().sum().sort_values(ascending=False).head(2)
```

```python
merge_df2['PREV_APP_COUNT'] = (
    merge_df2['PREV_APP_COUNT']
    .fillna(0)
    .astype(int)
)

merge_df2['APPROVAL_RATE'] = merge_df2['APPROVAL_RATE'].fillna(0)
```
